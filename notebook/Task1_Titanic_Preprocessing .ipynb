STEP 3: Handle Missing Values

What is Missing Value Handling?

👉 In real datasets, some values are empty (NaN)

👉 ML models cannot work with missing values

So we must:

Fill (Impute)
OR
Drop

Here the missing values are in the following columns.

1. Age - 177

2. cabin - 687

3. Embarked - 2.

Why Age has missing values?

    Some passengers’ age not recorded.

 Why NOT drop rows?

If you drop:

1. You may lose too much data.

2. Dataset becomes smaller.

Why use Median?

Ex:

Age values:

[22, 25, 30, 35, 100]

Mean = 42.4 ❌ (affected by outlier 100)

Median = 30 ✅ (middle value)


✅ Final Reason

👉 Median is robust to outliers

```python
# to fill the missing values of Age.
df_titanic['Age'] = df_titanic['Age'].fillna(df_titanic['Age'].median())
df_titanic.columns
```

Now: Handle Cabin Column.

Step 1: Check missing values


687 missing out of 891
🧠 Why Cabin is a problem?

👉 Around 77% data is missing

👉 That means:

Very little useful information

Filling will be inaccurate

🚫 Why NOT fill Cabin?

❌ Filling with mode/median?

Cabin is not numeric

Values are like: C85, E46 (not meaningful to average)

❌ Filling with random value?

Completely misleading

✅ Best Solution → DROP COLUMN

```python
df_titanic.drop('Cabin', axis=1, inplace=True)  # axis = 1 means column and inplace = True means modify original frame.
df_titanic.columns
```

Handle Embarked.

What is Embarked?

👉 It represents:

Port where passenger boarded
Values:
C → Cherbourg
Q → Queenstown
S → Southampton

👉 This is a categorical column

--->Only 2 missing values.

How to handle missing values?

✅ Use MODE (most frequent value)

Why mode?

Because:

Data is categorical

Cannot use mean or median

```python
df_titanic['Embarked'] = df_titanic['Embarked'].fillna(df_titanic['Embarked'].mode()[0])
```

```python
#after handling the missing value, we get,
df_titanic.isnull().sum()
```

step 4 convert data types


```python
df_titanic.info()
```

3. Pclass → int64 ❗

👉 Needs change:

Why important?

Even though values are numbers:

1, 2, 3 are NOT quantities

They are categories (class types)

👉 So:

❌ Wrong → treat as numeric

✅ Correct → treat as categorical

```python
df_titanic['Pclass'] = df_titanic['Pclass'].astype('category')
df_titanic['Sex'] = df_titanic['Sex'].astype('category')
df_titanic['Embarked'] = df_titanic['Embarked'].astype('category')
```

```python
df_titanic.info()
```

Step5 : Rename Columns

👉 Why rename?
Improve readability

Follow naming standards

Make code professional

🧠 Current column names

PassengerId, Survived, Pclass, SibSp, Parch

👉 Not very clean / consistent

🎯 Best Practice Naming

Use:

lowercase

underscore _

meaningful names

```python
df_titanic.rename(columns={
    'PassengerId': 'passenger_id',
    'Survived' : 'survival_status',
    'Pclass': 'passenger_class',
    'SibSp': 'sibling_spouse',
    'Parch': 'parent_child'

}, inplace = True)
df_titanic.columns
```
